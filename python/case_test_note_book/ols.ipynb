```python
print("SECTION 1: TESTS DE RÉGRESSION")

# Test 1: Simple 2D Linear Regression
print("\n--- Test 1: Régression Linéaire Simple 2D ---")
X1 = np.array([[1], [2]]).astype(np.float32)

Y1 = np.array([2, 3]).astype(np.float32)

# Convert for OLS (add bias term)
X1_bias = np.column_stack([np.ones(X1.shape[0]), X1]).astype(np.float32)
print(f"X avec bias: {X1_bias}")
print(f"Y: {Y1}")

# Training OLS
weights1 = bt.train_ols(X1_bias.tolist(), Y1.tolist())
print(f"Poids appris: {weights1}")

# Prediction
predictions1 = bt.predict_ols(X1_bias.tolist(), weights1)
print(f"Prédictions: {predictions1}")
print(f"Valeurs attendues: {Y1}")

# DataViz
plt.figure(figsize=(8, 6))
plt.scatter(X1.flatten(), Y1, color="blue", label="Données", s=100)
plt.scatter(
    X1.flatten(), predictions1, color="red", label="Prédictions OLS", s=100, marker="x"
)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Test 1: Régression Linéaire Simple 2D")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# SECTION 2: Tests de Classification
print("\n" + "=" * 50)
print("SECTION 2: TESTS DE CLASSIFICATION")
print("(Note: OLS n'est pas optimal pour la classification)")
print("=" * 50)

# Test 6: Simple Linear Classification
print("\n--- Test 6: Classification Linéaire Simple ---")
X6 = np.array([[1, 1], [2, 3], [3, 3]]).astype(np.float32)

Y6 = np.array([1, -1, -1]).astype(np.float32)

# Prepare Data for OLS
X6_bias = np.column_stack([np.ones(X6.shape[0]), X6]).astype(np.float32)
print(f"X avec bias: {X6_bias}")
print(f"Y: {Y6}")

# Training OLS
weights6 = bt.train_ols(X6_bias.tolist(), Y6.tolist())
print(f"Poids appris: {weights6}")

# Prediction
predictions6 = bt.predict_ols(X6_bias.tolist(), weights6)
print(f"Prédictions: {predictions6}")
print(f"Valeurs attendues: {Y6}")

# Apply threshold for classification
predictions6_class = [1 if p > 0 else -1 for p in predictions6]
print(f"Prédictions classifiées: {predictions6_class}")

# Viz
plt.figure(figsize=(8, 6))
plt.scatter(X6[Y6 == 1, 0], X6[Y6 == 1, 1], color="blue", label="Classe 1", s=100)
plt.scatter(X6[Y6 == -1, 0], X6[Y6 == -1, 1], color="red", label="Classe -1", s=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Test 6: Classification Linéaire Simple")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Test 7
print("\n--- Test 7: XOR (cas difficile pour OLS) ---")
X7 = np.array([[1, 0], [0, 1], [0, 0], [1, 1]]).astype(np.float32)
Y7 = np.array([1, 1, -1, -1]).astype(np.float32)

# Prepare Data for OLS
X7_bias = np.column_stack([np.ones(X7.shape[0]), X7]).astype(np.float32)
print(f"X avec bias: {X7_bias}")
print(f"Y: {Y7}")

# training OLS
weights7 = bt.train_ols(X7_bias.tolist(), Y7.tolist())
print(f"Poids appris: {weights7}")

# Prediction
predictions7 = bt.predict_ols(X7_bias.tolist(), weights7)
print(f"Prédictions: {predictions7}")
print(f"Valeurs attendues: {Y7}")

# Apply classification threshold
predictions7_class = [1 if p > 0 else -1 for p in predictions7]
print(f"Prédictions classifiées: {predictions7_class}")

# Viz
plt.figure(figsize=(8, 6))
plt.scatter(X7[Y7 == 1, 0], X7[Y7 == 1, 1], color="blue", label="Classe 1", s=100)
plt.scatter(X7[Y7 == -1, 0], X7[Y7 == -1, 1], color="red", label="Classe -1", s=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Test 7: XOR (OLS devrait échouer)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# SECTION 3: Calcul des métriques d'évaluation

print("\n" + "=" * 50)
print("SECTION 3: ÉVALUATION DES PERFORMANCES")
print("=" * 50)


def calculate_mse(y_true, y_pred):
    """Calcule l'erreur quadratique moyenne"""
    return np.mean((np.array(y_true) - np.array(y_pred)) ** 2)


def calculate_mae(y_true, y_pred):
    """Calcule l'erreur absolue moyenne"""
    return np.mean(np.abs(np.array(y_true) - np.array(y_pred)))


def calculate_r2(y_true, y_pred):
    """Calcule le coefficient de détermination R²"""
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)


# Évaluation des tests de régression
tests_regression = [
    ("Test 1: Régression Simple 2D", Y1, predictions1),
    ("Test 2: Régression Non-Linéaire 2D", Y2, predictions2),
    ("Test 3: Régression Simple 3D", Y3, predictions3),
    # ("Test 4: Régression Tricky 3D", Y4, predictions4),
    ("Test 5: Régression Non-Linéaire 3D", Y5, predictions5),
]

print("\n--- Métriques de Régression ---")
for test_name, y_true, y_pred in tests_regression:
    mse = calculate_mse(y_true, y_pred)
    mae = calculate_mae(y_true, y_pred)
    r2 = calculate_r2(y_true, y_pred)
    print(f"{test_name}:")
    print(f"  MSE: {mse:.4f}")
    print(f"  MAE: {mae:.4f}")
    print(f"  R²:  {r2:.4f}")


# Évaluation des tests de classification
def calculate_accuracy(y_true, y_pred):
    """Calcule la précision de classification"""
    return np.mean(np.array(y_true) == np.array(y_pred))


tests_classification = [
    ("Test 6: Classification Simple", Y6, predictions6_class),
    ("Test 7: XOR", Y7, predictions7_class),
]

print("--- Métriques de Classification ---")
for test_name, y_true, y_pred in tests_classification:
    accuracy = calculate_accuracy(y_true, y_pred)
    print(f"{test_name}:")
    print(f"  Précision: {accuracy:.4f}")
    print()

print("=" * 50)
print("TESTS TERMINÉS")
print("=" * 50)
```
