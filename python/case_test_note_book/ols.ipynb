```python
print("SECTION 1: TESTS DE RÉGRESSION")

# Test 1: Simple 2D Linear Regression
print("\n--- Test 1: Régression Linéaire Simple 2D ---")
X1 = np.array([
    [1],
    [2]
]).astype(np.float32)

Y1 = np.array([2, 3]).astype(np.float32)

# Convert for OLS (add bias term)
X1_bias = np.column_stack([np.ones(X1.shape[0]), X1]).astype(np.float32)
print(f"X avec bias: {X1_bias}")
print(f"Y: {Y1}")

# Training OLS
weights1 = bt.train_ols(X1_bias.tolist(), Y1.tolist())
print(f"Poids appris: {weights1}")

# Prediction
predictions1 = bt.predict_ols(X1_bias.tolist(), weights1)
print(f"Prédictions: {predictions1}")
print(f"Valeurs attendues: {Y1}")

# DataViz
plt.figure(figsize=(8, 6))
plt.scatter(X1.flatten(), Y1, color='blue', label='Données', s=100)
plt.scatter(X1.flatten(), predictions1, color='red', label='Prédictions OLS', s=100, marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Test 1: Régression Linéaire Simple 2D')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Test 10: Multi-class Classification (3 classes using multiple OLS)
print("\n--- Test 10: Classification Multi-classe (3 classes) ---")
np.random.seed(42)
X10 = np.random.random((500, 2)) * 2.0 - 1.0

# Create 3 classes based on linear separators
Y10_multi = np.array([
    [1, 0, 0] if -p[0] - p[1] - 0.5 > 0 and p[1] < 0 and p[0] - p[1] - 0.5 < 0 else
    [0, 1, 0] if -p[0] - p[1] - 0.5 < 0 and p[1] > 0 and p[0] - p[1] - 0.5 < 0 else
    [0, 0, 1] if -p[0] - p[1] - 0.5 < 0 and p[1] < 0 and p[0] - p[1] - 0.5 > 0 else
    [0, 0, 0] for p in X10
])

# Filter out points that don't belong to any class
valid_mask = ~np.all(Y10_multi == [0, 0, 0], axis=1)
X10 = X10[valid_mask].astype(np.float32)
Y10_multi = Y10_multi[valid_mask]

print(f"X shape after filtering: {X10.shape}")
print(f"Y shape: {Y10_multi.shape}")

# Prepare data for OLS
X10_bias = np.column_stack([np.ones(X10.shape[0]), X10]).astype(np.float32)

# Train separate OLS models for each class (One-vs-Rest)
weights10_class1 = bt.train_ols(X10_bias.tolist(), Y10_multi[:, 0].tolist())
weights10_class2 = bt.train_ols(X10_bias.tolist(), Y10_multi[:, 1].tolist())
weights10_class3 = bt.train_ols(X10_bias.tolist(), Y10_multi[:, 2].tolist())

print(f"Poids classe 1: {weights10_class1}")
print(f"Poids classe 2: {weights10_class2}")
print(f"Poids classe 3: {weights10_class3}")

# Predictions for each class
pred10_class1 = bt.predict_ols(X10_bias.tolist(), weights10_class1)
pred10_class2 = bt.predict_ols(X10_bias.tolist(), weights10_class2)
pred10_class3 = bt.predict_ols(X10_bias.tolist(), weights10_class3)

# Combine predictions (argmax)
predictions10_combined = []
for i in range(len(pred10_class1)):
    scores = [pred10_class1[i], pred10_class2[i], pred10_class3[i]]
    predicted_class = np.argmax(scores)
    predictions10_combined.append(predicted_class)

# Convert true labels to class indices
Y10_true_classes = np.argmax(Y10_multi, axis=1)

# Visualisation
plt.figure(figsize=(12, 5))

# Plot 1: True classes
plt.subplot(1, 2, 1)
class1_mask = Y10_true_classes == 0
class2_mask = Y10_true_classes == 1
class3_mask = Y10_true_classes == 2

plt.scatter(X10[class1_mask, 0], X10[class1_mask, 1], color='blue', label='Classe 1', alpha=0.6, s=20)
plt.scatter(X10[class2_mask, 0], X10[class2_mask, 1], color='red', label='Classe 2', alpha=0.6, s=20)
plt.scatter(X10[class3_mask, 0], X10[class3_mask, 1], color='green', label='Classe 3', alpha=0.6, s=20)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Classes vraies')
plt.legend()
plt.grid(True)

# Plot 2: Predicted classes
plt.subplot(1, 2, 2)
pred_class1_mask = np.array(predictions10_combined) == 0
pred_class2_mask = np.array(predictions10_combined) == 1
pred_class3_mask = np.array(predictions10_combined) == 2

plt.scatter(X10[pred_class1_mask, 0], X10[pred_class1_mask, 1], color='blue', label='Classe 1', alpha=0.6, s=20)
plt.scatter(X10[pred_class2_mask, 0], X10[pred_class2_mask, 1], color='red', label='Classe 2', alpha=0.6, s=20)
plt.scatter(X10[pred_class3_mask, 0], X10[pred_class3_mask, 1], color='green', label='Classe 3', alpha=0.6, s=20)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Classes prédites')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculate accuracy
accuracy10 = calculate_accuracy(Y10_true_classes, predictions10_combined)
print(f"Précision multi-classe: {accuracy10:.4f}")
```

```python
# SECTION 3: Calcul des métriques d'évaluation

print("\n" + "="*50)
print("SECTION 3: ÉVALUATION DES PERFORMANCES")
print("="*50)

def calculate_mse(y_true, y_pred):
    """Calcule l'erreur quadratique moyenne"""
    return np.mean((np.array(y_true) - np.array(y_pred)) ** 2)

def calculate_mae(y_true, y_pred):
    """Calcule l'erreur absolue moyenne"""
    return np.mean(np.abs(np.array(y_true) - np.array(y_pred)))

def calculate_r2(y_true, y_pred):
    """Calcule le coefficient de détermination R²"""
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

# Évaluation des tests de régression
tests_regression = [
    ("Test 1: Régression Simple 2D", Y1, predictions1),
    ("Test 2: Régression Non-Linéaire 2D", Y2, predictions2),
    ("Test 3: Régression Simple 3D", Y3, predictions3),
   # ("Test 4: Régression Tricky 3D", Y4, predictions4),
    ("Test 5: Régression Non-Linéaire 3D", Y5, predictions5)
]

print("\n--- Métriques de Régression ---")
for test_name, y_true, y_pred in tests_regression:
    mse = calculate_mse(y_true, y_pred)
    mae = calculate_mae(y_true, y_pred)
    r2 = calculate_r2(y_true, y_pred)
    print(f"{test_name}:")
    print(f"  MSE: {mse:.4f}")
    print(f"  MAE: {mae:.4f}")
    print(f"  R²:  {r2:.4f}")


# Évaluation des tests de classification
def calculate_accuracy(y_true, y_pred):
    """Calcule la précision de classification"""
    return np.mean(np.array(y_true) == np.array(y_pred))

tests_classification = [
    ("Test 6: Classification Simple", Y6, predictions6_class),
    ("Test 7: XOR", Y7, predictions7_class)
]

print("--- Métriques de Classification ---")
for test_name, y_true, y_pred in tests_classification:
    accuracy = calculate_accuracy(y_true, y_pred)
    print(f"{test_name}:")
    print(f"  Précision: {accuracy:.4f}")
    print()

print("="*50)
print("TESTS TERMINÉS")
print("="*50)
```
